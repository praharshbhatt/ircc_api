```python
import requests
import pandas as pd

# URL of the IRCC JSON endpoint
url = "https://www.canada.ca/content/dam/ircc/documents/json/ee_rounds_123_en.json?wbCacheBust=9e127c23-64ec-43cc-8bba-d2f0a26de76b"

# Fetching data from the URL
response = requests.get(url)
data = response.json()

# Extracting relevant information from the JSON data
rounds_data = []
for round_info in data["rounds"]:
    round_number = round_info["drawNumber"]
    draw_date = round_info["drawDateFull"]
    round_type = round_info["drawName"]
    invitations_issued = round_info["drawSize"]
    crs_score = round_info["drawCRS"]
    rounds_data.append([round_number, draw_date, round_type, invitations_issued, crs_score])

# Creating a DataFrame from the extracted data
df = pd.DataFrame(rounds_data, columns=["Round", "Date", "Round type", "Invitations issued", "CRS score"])

# Displaying the DataFrame
df
```
